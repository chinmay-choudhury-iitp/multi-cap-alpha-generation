```python
import yfinance as yf
import pandas as pd
from google.colab import files

tickers = {
    # Your 10 portfolio stocks
    "ICICI_AMC":     "ICICIPRULI.NS",
    "Force_Motors":  "FORCEMOT.NS",
    "Garden_Reach":  "GRSE.NS",
    "Waaree":        "WAAREEENER.NS",
    "Tips_Music":    "TIPSMUSIC.NS",
    "Central_Mine":  "CMPDI.NS",
    "Shilchar":      "SHILCTECH.NS",
    "Swaraj":        "SWARAJENG.NS",
    "Sika":          "SIKA.NS",
    "Monarch":       "MONARCH.NS",

    # 3 proxy stocks for sparse-data replacements
    "Proxy_Vinati":  "VINATIORGA.NS",   # for Sika Interplant
    "Proxy_NMDC":    "NMDC.NS",         # for CMPDI
    "Proxy_TARIL":   "TARIL.NS",        # for Shilchar
}

start = "2025-05-20"
end   = "2026-05-20"

prices = pd.DataFrame()

for name, ticker in tickers.items():
    data = yf.download(ticker, start=start, end=end, auto_adjust=True, progress=False)
    if data.empty:
        print(f"WARNING: No data for {name} ({ticker}) — check ticker symbol")
    else:
        prices[name] = data["Close"]
        print(f"OK: {name} — {len(data)} rows fetched")

prices.to_excel("stock_prices.xlsx", index=True)
print("\nSaved: stock_prices.xlsx")

# This line downloads it to your computer
files.download("stock_prices.xlsx")
```

```python
import pandas as pd

xl = pd.read_excel('stock_prices.xlsx', index_col=0)

# Sika's own 21-day returns and sigma
sika_returns = xl['Sika'].dropna().pct_change().dropna()
sika_sigma = sika_returns.std()

print(f"Sika 21-day σ: {sika_sigma:.6f}")
print(f"Days used: {len(sika_returns)}")

# Compare with Vinati proxy
vinati_returns = xl['Proxy_Vinati'].dropna().pct_change().dropna()
print(f"Vinati proxy σ: {vinati_returns.std():.6f}")
```
